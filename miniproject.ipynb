```python
# Simple Calculator

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)

class Calculator:
    def add(self, a, b):
        return a + b

    def sub(self, a, b):
        return a - b

    def mul(self, a, b):
        return a * b

    def div(self, a, b):
        return a / b

c = Calculator()
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

print(c.add(x, y))
print(c.sub(x, y))
print(c.mul(x, y))
print(c.div(x, y)) 
```
